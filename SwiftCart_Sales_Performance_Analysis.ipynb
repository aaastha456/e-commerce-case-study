```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('ecommerce_sales_dataset.csv')
print("Data Loaded Successfully!")
df.head()
```

```python
# Task 1: Total_Sales = Price * Quantity
df['Total_Sales'] = df['Price'] * df['Quantity']
print("Revenue Calculated!")
df[['Product_Category', 'Price', 'Quantity', 'Total_Sales']].head()
```

```python
# Task 2: Grouping by Category
category_sales = df.groupby('Product_Category')['Total_Sales'].sum().sort_values(ascending=False)
print("Sales by Category:")
print(category_sales)

top_cat = category_sales.idxmax()
print(f"\nTop Performing Category (Cash Cow): {top_cat}")
```

```python
# Task 3: Average Rating
avg_rating = df['Customer_Rating'].mean()
print(f"Overall Average Customer Rating: {avg_rating:.2f}")

# Interpretation logic
if avg_rating > 4:
    print("Interpretation: High Customer Satisfaction.")
else:
    print("Interpretation: Moderate Satisfaction, scope for improvement.")
```

```python
# Task 4: Filter Rating > 4
premium_products = df[df['Customer_Rating'] > 4]
print(f"Found {len(premium_products)} Premium Products:")
premium_products[['Order_ID', 'Product_Category', 'Customer_Rating']]
```

```python
# Task 5: Category vs Total Sales Chart
plt.figure(figsize=(10,6))
category_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Sales Performance by Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Task 6: Price Distribution
plt.figure(figsize=(10,6))
plt.hist(df['Price'], bins=8, color='salmon', edgecolor='black')
plt.title('Analysis of Product Price Distribution')
plt.xlabel('Price ($)')
plt.ylabel('Number of Products')
plt.show()
```
